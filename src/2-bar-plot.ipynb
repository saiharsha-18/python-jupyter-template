### Exercise 2: Bar Plot with Colors

```python
# TASK: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 9].
# Use different colors for each bar and add a title to the plot.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['r', 'g', 'b', 'y']  # Different colors for each bar

# Create the bar plot
plt.figure(figsize=(6, 4))
plt.bar(categories, values, color=colors)

# Customizations
plt.title("Category Value Distribution")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Show plot
plt.show()
```
